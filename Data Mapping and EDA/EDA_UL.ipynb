## Combine

```python
df_all = pd.concat([df_2021, df_2011, df_2001], ignore_index=True)
df_all = df_all.fillna(0)
df_all
```

```python
df_all_1 = df_all.groupby(['Town','Year'])[["Total: Highest Qualification", \
                                            "No qualifications", \
                                            "Level 1 qualifications", \
                                            "Level 2 qualifications", \
                                            "Apprenticeship", \
                                            "Level 3 qualifications", \
                                            "Level 4 qualifications or above", \
                                            "Other qualifications", \
                                            "Total: Economic activity status", \
                                            "Economically active:In employment", \
                                            "Economically active: Unemployed", \
                                            "Economically inactive"]].apply(lambda x : x.astype(int).sum()).reset_index()
df_all_1['No qualifications%'] = df_all_1['No qualifications'] / df_all_1['Total: Highest Qualification']
df_all_1['Level 1 qualifications%'] = df_all_1['Level 1 qualifications'] / df_all_1['Total: Highest Qualification']
df_all_1['Level 2 qualifications%'] = df_all_1['Level 2 qualifications'] / df_all_1['Total: Highest Qualification']
df_all_1['Apprenticeship%'] = df_all_1['Apprenticeship'] / df_all_1['Total: Highest Qualification']
df_all_1['Level 4 qualifications or above%'] = df_all_1['Level 4 qualifications or above'] / df_all_1['Total: Highest Qualification']
df_all_1['Other qualifications%'] = df_all_1['Other qualifications'] / df_all_1['Total: Highest Qualification']

df_all_1['Economically active:In employment%'] = df_all_1['Economically active:In employment'] / df_all_1['Total: Economic activity status']
df_all_1['Economically active: Unemployed%'] = df_all_1['Economically active: Unemployed'] / df_all_1['Total: Economic activity status']
df_all_1['Economically inactives%'] = df_all_1['Economically inactive'] / df_all_1['Total: Economic activity status']

df_all_1
```

```python
df_all.to_csv('raw_data.csv', index=False)  
df_all_1.to_csv('output.csv', index=False)  
```
